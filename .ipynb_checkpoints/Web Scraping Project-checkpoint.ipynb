```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
import requests
```

```python
content = requests.get(url).text
```

```python
type(content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
tb = soup.find('tbody')
```

```python
tb
```

```python
table_row = tb.find_all('tr')
```

```python
table_row
```

```python
table_row[8]
```

```python
all_data = []
for items in range(8, len(table_row)):
    country_data = []
    tr = table_row[items]
    td = tr.find_all('td')
    for dp in td: # scraps each country data
        country_data.append(dp.text)
    all_data.append(country_data) # appends each country data in whole data
```

```python
all_data
```

```python
all_data[0]
```

```python
# 1:10 14
```

```python
filtered_data = list(map(lambda x: x[1:10] + [x[14]], all_data))
```

```python
filtered_data
```

```python
column_names = ['Country', 'Total Cases', 'New Cases', 'Total Deaths', 'New Deaths', 'Total Recovered', 'New Recovered', 'Active Cases', 'Serious/Critical', 'Population']
```

```python
for i in column_names:
    print(i)
```

```python
import pandas as pd
```

```python
pd.DataFrame(filtered_data)
```

```python
df = pd.DataFrame(filtered_data, columns = column_names)
```

```python
df.to_csv('covid-jan2.csv', index = False)
```
